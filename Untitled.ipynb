```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')  # Replace with your actual dataset path if different
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Survived'].value_counts()
df['Pclass'].value_counts()
df['Sex'].value_counts()
```

```python
#Pairplot: Survivors are more frequent in higher classes and younger ages.
sns.pairplot(df[['Survived','Pclass','Age','Fare']])
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['Age'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(False)
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=df, hue='Pclass', palette='pastel', legend=False)
plt.title('Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['Age'], df['Fare'], alpha=0.6, edgecolors='w', color='teal')
plt.title('Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```
